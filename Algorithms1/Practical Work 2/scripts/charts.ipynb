```python
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 8, 8
plt.savefig('foo.png')
```

```python
x = [5,6,10,100,123,256]
y1 = [55,34.8,146.6,6990,11466.2,34468.6]
y2 = [4.1833,2.3875,15.3558,981.8915,1089.0905,1227.6888]
```

```python
plt.plot(x,y1,label="Averages")
plt.plot(x,y2,label="standard Deviation")

plt.title('dynamic approach')
plt.ylabel('Time of execution(microseconds)',fontsize=14)
plt.xlabel('size of array',fontsize=14)


plt.legend()

plt.savefig('foo.png', format='png', dpi=1200)
plt.show()
plt.close()
```
