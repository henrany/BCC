```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support

from sklearn.neighbors import KNeighborsClassifier

# Um fator importante é que o SKLearn não cria conjuntos de validação para você. Você tem algumas abordagens,
# uma é realizar um novo split no treino. Outra é fazer uso de classificadores com CV no fim.
# Tipo LogisticRegressionCV (ver na API). Por fim, você pode fazer uso da classe GridSearchCV.
# Leia a documentação da mesma.
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import numpy as np
```

```python
#getting our dataset
df = pd.read_csv("winequality-white.csv", sep = ";")
df
```

```python
#we took 20% of the data for test and 80% for the training
y = df['quality']
x_train,x_test,y_train,y_test = train_test_split(df,y,test_size=0.2)
```

```python
x_test
```

```python
#perfoming logistics regression on our data 
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
#we now look at getting the coefficient and intercept of our model
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
#checking how our model is performing
lr_pred = lr.predict(x_test)
```

```python
#to perform validation using knn, we take our k value to be 3
classifier = KNeighborsClassifier(n_neighbors=3,metric='euclidean')
classifier.fit(x_train, y_train)
```

```python
#checking how our model performs 
k_pred = classifier.predict(x_test) 
```

```python
#getting our precisions 
knn_precision = precision_recall_fscore_support(y_test, k_pred, average='weighted')
lr_precision = precision_recall_fscore_support(y_test, lr_pred, average='weighted')
```

```python
print("Quality of knn model")
print("Precision:", knn_precision[0])
print("Recall:", knn_precision[1])
print("F1:", knn_precision[2], '\n')


print("Quality of regression model")
print("Precision:", lr_precision[0])
print("Recall:", lr_precision[1])
print("F1:", lr_precision[2])
```

```python
#now we get our confusion matrix
lr_conf_mat = confusion_matrix(y_test, lr_pred)
knn_conf_mat = confusion_matrix(y_test, k_pred)
```

```python
#printing the confusion matrix
print("confusion matrix of logistic regression")
print(lr_conf_mat)
```

```python
#printing the confusion matrix
print("confusion matrix of knn")
print(knn_conf_mat)
```
